# Network Construction

```python
import torch
import torch.nn as nn #객체로 추상화
import torch.nn.functional as F #함수로 추상화

class ClassificationModel(nn.Module) :
  def __init__(self) :
    super(ClassificationModel, self).__init__()

    #Define convolution operaion & Pooling
    #input 채널이 3일때 32이로 바꾸고자할때 convolution 사용하는 방법
    self.conv1 = nn.Conv2d(in_channels=3, out_channels=32, kernel_size=3, stride=1, padding=1)

    #Maxpooling을 사용해서 kernel 안에 있는 가장 큰 값을 뽑아 return(2*2 -> 1*1로 변환하기 때문에 64->32로 변함)
    self.pool1 = nn.MaxPool2d(kernel_size=2, stride =2)

    self.conv2 = nn.Conv2d(in_channels=32, out_channels=64, kernel_size=3, stride=1, padding=1)
    self.pool2 = nn.MaxPool2d(kernel_size=2, stride=2)

    self.linear = nn.Linear(16*16*64, 1) #입력 : 16*16*64, 결과 : 개, 고양이 선택 - 1

  def forward(self, x) :
    x = F.relu(self.conv1(x))
    x = self.pool1(x)
    x = F.relu(self.conv2(x))
    x = self.pool2(x)
    x = x.view(-1, 16*16*64)
    x = torch.sigmoid(self.linear(x))
    return x

import time

net = ClassificationModel().cuda()
s = time.time()

for i in range(3):
  dummy_input = torch.rand(50, 3, 64, 64).cuda()
  out = net(dummy_input)
  print(out.shape)

e = time.time()
print(e-s)


# forward x에대한 설명
"""
#F.relu - activation 사용 이유 : 값의 range를 정하기 위해
conv1_out = F.relu(conv1(dummy_input))
print('input shape :', dummy_input.shape, 'output shape:', conv1_out.shape)
pool1_out = pool1(conv1_out)
print('input shape :', conv1_out.shape, 'output shape:', pool1_out.shape)

conv2_out = F.relu(conv2(pool1_out))
print('input shape :', pool1_out.shape, 'output shape:', conv2_out.shape)
pool2_out = pool2(conv2_out)
print('input shape :', conv2_out.shape, 'output shape:', pool2_out.shape)

#모든 feature와 모든 output 채널이 다 연결 -> 모든 데이터를 하나의 feature로 본다
hidden = pool2_out.view(-1, 16*16*64)
print(hidden.shape)
#sigmoid를 사용해 0-0.5 면 고양이, 0.5-1면 개로 판단하도록 설정함
out = torch.sigmoid(linear(hidden))
print(out.shape) # scalar 값이 나와서 개인지 고양인지를 알아낼 수 있음
"""
```

# Network Training

```python
import torch
import torch.nn as nn
import torch.utils.data as data
import os
import matplotlib.pyplot as plt

print('train dataset length: ', len(train_dataloader)) #2만장 이미지 -> 4개씩 가져와서 5000
print('test dataset length: ', len(test_dataloader)) # 2500개 이미지 -> 4개씩 가져와서 625

# 네트워크 세팅
net = ClassificationModel().cuda()

# loss(criterion), optimizer 세팅
import torch.optim as optim
criterion = nn.BCELoss() #loss랑 동일
optimizer = torch.optim.Adam(net.parameters(), lr=0.0001)

# 기타 변수
train_info = []
val_info = []
object_epoch = 20

save_path = './ClassificationNetwork'
os.makedirs(save_path, exist_ok = True)
output_path = os.path.join(save_path, 'basic_model.tar')
```

```python
def train_1epoch(net, dataloader):
  confusion_matrix = [[0,0], [0,0]]
  total_loss = 0
  iteration = 0

  net.train()

  for sample in tqdm.auto.tqdm(dataloader) :
    img = sample['img']
    label = sample['label']

    img = img.float().cuda()
    label = label.float().cuda()

    optimizer.zero_grad()
    output = net(img).squeeze(1)

    loss = criterion(output, label)
    loss.backward()
    optimizer.step()

    total_loss += loss.detach()
    iteration += 1

    for i in range(len(label)) :
      real_class = int(label[i])
      pred_class = int(output[i]>0.5)
      confusion_matrix[real_class][pred_class] += 1
  
  positive = confusion_matrix[0][0] + confusion_matrix[1][1]
  negative = confusion_matrix[1][0] + confusion_matrix[0][1]
  accuracy = positive / (positive + negative)
  total_loss /= iteration

  return accuracy, total_loss

def validation_epoch(net, dataloader):
  confusion_matrix = [[0,0], [0,0]]
  total_loss = 0
  iteration = 0

  net.eval()

  for sample in tqdm.auto.tqdm(dataloader) :
    img = sample['img']
    label = sample['label']

    img = img.float().cuda()
    label = label.float().cuda()

    output = net(img).squeeze(1)

    loss = criterion(output, label)

    total_loss += loss.detach()
    iteration += 1

    for i in range(len(label)) :
      real_class = int(label[i])
      pred_class = int(output[i]>0.5)
      confusion_matrix[real_class][pred_class] += 1
  
  positive = confusion_matrix[0][0] + confusion_matrix[1][1]
  negative = confusion_matrix[1][0] + confusion_matrix[0][1]
  accuracy = positive / (positive + negative)
  total_loss /= iteration

  return accuracy, total_loss

best_accuracy = 0
for epoch in range(object_epoch) :
  train_accuracy, train_loss = train_1epoch(net,train_dataloader)
  print('[TRANING] Epoch {} : accuracy: {}, loss: {}'.format(epoch, train_accuracy, train_loss))
  train_info.append({'loss': train_loss, 'accuracy' : train_accuracy})

  with torch.no_grad():
    val_accuracy, val_loss = validation_1epoch(net, val_dataloader)
  print('[VALIDATION] Epoch {} : accuracy: {}, loss: {}'.format(epoch, val_accuracy, val_loss))
  val_info.append({'loss': val_loss, 'accuracy' : val_accuracy})

  if best_accuracy < val_accuracy :
    best_accuracy = val_accuracy

    torch.save({
        'memo' : 'Classification Model!!',
        'accuracy' : val_accuracy,
        'model_weight' : net.state_dict()
    }, output_path)

    torch.save({
        'memo' : 'Classification Model!!',
        'accuracy' : val_accuracy,
        'model_weight' : net.state_dict()
    }, save_path + "/validation_model.tar")
```
